```python
from cleanup import fade_end, detrend, normalize
from reformat import to_f32, to_mono, resample, trim_pad

from fs import file_loader, file_storer, label
from functional import seq


stream = (
    seq(file_loader('../_temp/'))
    .map(to_f32)
    .map(to_mono)
    .map(resample)
    .map(lambda f: trim_pad(f, 128 * 512))
    .map(lambda f: fade_end(f, 1000))
    .map(detrend)
    .map(normalize)
    .map(lambda f: label(f, 'stylized'))
)
file_storer('../_temp/clean', stream)
```
