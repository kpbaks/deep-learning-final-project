```python
from fs import AudioFile, file_loader, file_storer, label
from cleanup import *
from reformat import *
from pathlib import Path
import matplotlib.pyplot as plt
from functional import seq

stream = (seq(file_loader("../_temp/"))
    .map(to_f32)
    .map(to_mono)
    .map(resample)
    .map(lambda f: trim_pad(f, 128 * 512))
    .map(lambda f: fade_end(f, 1000))
    .map(detrend)
    .map(normalize)
    .map(lambda f: label(f, "subtle"))
)
file_storer("../_temp/clean", stream)
```
